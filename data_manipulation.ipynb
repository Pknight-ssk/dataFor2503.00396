```python
import numpy as np
import h5py
from matplotlib import pyplot as plt


def get_avg_loop_length_planar(filename):
    with h5py.File(filename, "r") as f:
        
        avg_loop_length = np.zeros((49,49))
        for i in range(1,50):
            for j in range(1,50):
                data_lea = f[f"log_elements_array_{i}_{j}"][()] 
                data_ca = f[f"counts_array_{i}_{j}"][()] 
                avg_loop_length[i-1,j-1] = np.log2(np.sum(data_ca*np.exp2(data_lea))/np.sum(data_ca))
    return avg_loop_length

avg_ll_11_uniform = get_avg_loop_length_planar("loop_circuit_sample_11_10_0d02_p_q.hdf5")
#  ()
x = np.arange(0.02, 1, 0.02)
y = np.arange(0.02, 1, 0.02)
X, Y = np.meshgrid(x, y)
Z = avg_ll_11_uniform

#  X 
Y_prime = 0.5 + (Y - 0.5) * (1 - X)

# 
fig, ax = plt.subplots(figsize=(8, 6))
pcm = ax.pcolormesh(Y_prime, X, Z)
# pcm = ax.pcolormesh(Y, X, Z)
plt.colorbar(pcm, label='Log(avg. loop length)')
ax.set_ylim(0, 1)
ax.set_xlabel("q'")
ax.set_ylabel('p')
plt.show()
```
