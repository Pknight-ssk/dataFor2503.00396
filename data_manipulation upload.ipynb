```python
import numpy as np
import h5py
from matplotlib import pyplot as plt

planar_data11 = np.zeros((49,49))
with h5py.File("loop_circuit_sample_11_10_0d02_p_q.hdf5", "r") as f:
    for i in range(1,50):
        for j in range(1,50):
            planar_data11[i-1, j-1] = f[f"avg_spanning_num_{i}_{j}"][()]

plt.style.use('_mpl-gallery-nogrid')


x = np.arange(0.02, 1, 0.02)
y = np.arange(0.02, 1, 0.02)
X, Y = np.meshgrid(x, y)
Z = planar_data11


Y_prime = 0.5 + (Y - 0.5) * (1 - X)


fig, ax = plt.subplots(figsize=(8, 6))
pcm = ax.pcolormesh(Y_prime, X, Z)
cbar = plt.colorbar(pcm, label='Log(Spanning Number)')


ax.set_xlabel("q'", fontsize=24)
ax.set_ylabel('p', fontsize=24)


ax.tick_params(axis='both', which='major', labelsize=18)  
ax.tick_params(axis='both', which='minor', labelsize=14)  


cbar.ax.tick_params(labelsize=18)  
cbar.set_label('Log(S)', fontsize=24)  

ax.set_ylim(0, 1)

plt.show()

```
